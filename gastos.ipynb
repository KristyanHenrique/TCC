```python
import pandas as pd
import requests
```

```python
listaParlamentar = pd.read_csv("./dados/ListaParlamentarEmExercicio.csv", sep=";" )
listaParlamentar = listaParlamentar.groupby(['CodigoParlamentar']).first()
```

```python
def requisitarComissoesDoSenador(codigo,nome):
    print(codigo,nome)
    tabela = pd.DataFrame()
    for i in range(2021,2025):
        df = pd.read_html('https://www6g.senado.leg.br/transparencia/sen/'+str(codigo)+'/?ano='+str(i))
        df0 = df[0].T
        df0.drop(index=("Recurso"),inplace=True)
        df0["ano"] = i
        df0["CodigoParlamentar"] = codigo
        df0 = df0.rename(columns={0 : "AluguelEscritorio", 1 : "MateriaisConsumo", 2: "HospedagemCombustivel", 3:"ApoioParlamentar", 4:"Propaganda", 5:"Passagens",6:"SegurançaPrivada",7:"Total", "ano":"ano","CodigoParlamentar":"CodigoParlamentar"})
        df1 = df[1].T
        df2 = df[2].T
        df1.drop(index=("Recurso"),inplace=True)
        df2.drop(index=("Benefício"),inplace=True)
        df0["Viagens"] = df1[0]
        df0["Diarias"] = df1[1]
        df0["RessarcimentePassagens"] = df1[2]
        df0["Passagens"] = df1[3]
        df0["ConsumoMaterial"] = df1[4]
        df0["Combustiveis"] = df1[5]
        df0["OutrosMateriais"] = df1[6]
        df0["Correios"] = df1[7]
        df0["MarketingDigital"] = df1[8]
        df0["PassagensAereas"] = df1[9]
        df0["Total"] = df1[10]
        tabela = pd.concat([tabela, df0], axis=0)
    tabela.to_csv(f'./dados/gastos/{nome}.csv',index=False)
```

```python
for i,r in listaParlamentar.iterrows():
    requisitarComissoesDoSenador(i,r['NomeParlamentar'])
```
