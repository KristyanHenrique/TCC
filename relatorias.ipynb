```python
import pandas as pd
import requests
import xmltodict
import json
import csv
import xml.etree.ElementTree as ET
```

```python
listaParlamentar = pd.read_csv("./dados/ListaParlamentarEmExercicio.csv", sep=";" )
listaParlamentar = listaParlamentar.groupby(['CodigoParlamentar']).first()
```

```python
def requisitarApartesDoSenador(codigo,nome):
    print(nome, codigo)
    url = 'https://legis.senado.leg.br/dadosabertos/senador/'+str(codigo)+'/relatorias'
    response = requests.get(url)
    tabela = []
    if response.status_code == 200:
        tree = ET.fromstring(response.content)
        for elemento in tree.findall('.//Relatoria'):
            try:
                TipoRelator = elemento.find('DescricaoTipoRelator').text
            except:
                TipoRelator = "N/I"
            try:
                DataDesignacao = elemento.find('DataDesignacao').text
            except:
                DataDesignacao = "N/I"
            Identificacao = elemento.find('Materia/DescricaoIdentificacao').text
            Ementa = elemento.find('Materia/Ementa').text
            AnoMateria = elemento.find('Materia/Ano').text
            SiglaComissao = elemento.find('Comissao/Sigla').text
            NomeComissao = elemento.find('Comissao/Nome').text
            tabela.append([codigo,TipoRelator,DataDesignacao,Identificacao,Ementa,AnoMateria,SiglaComissao,NomeComissao,])
        df = pd.DataFrame(data=tabela, columns=['codigoParlamentar','TipoRelator','DataDesignacao','Identificacao','Ementa','AnoMateria','SiglaComissao','NomeComissao'])
        df.to_csv(f'./dados/relatorias/{nome}.csv',index=False)
    else:
        print("deu Ruim")
```

```python
for i,r in listaParlamentar.iterrows():
    requisitarApartesDoSenador(i,r['NomeParlamentar'])
```
